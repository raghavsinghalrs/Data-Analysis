# Q.2 Which sport is having most number of Gold Medals so far?

```python
data=df[df['Medal'] == 'Gold']
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Sport',data=data)
plt.xticks(rotation=90)
```

```python
# From this graph we can know that the sport which having most number of gold medals are:
print("Top 5 are: ")
print("1. Aquatics ")
print("2. Athletics ")
print("3. Rowing ")
print("4. Gymnastics ")
print("5. Fencing ")
```

# Q.3 Which sport is having most number of medals so far?

```python
bucket=[]
for i in df['Sport'].unique():
    bucket.append([i,len(df[df['Sport']==i])])
bucket = pd.DataFrame(bucket,columns=['Sport','Medal'])
bucket = bucket.sort_values(by='Medal',ascending=False).head()
print("Top 5 sports that have most number of medals: ")
print(bucket)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Sport',y='Medal',data=bucket)
plt.xticks(rotation=45)
```
