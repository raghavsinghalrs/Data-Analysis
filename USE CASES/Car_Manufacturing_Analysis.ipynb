# Top 5  Car brands in terms of number of cars

```python
plt.Figure(figsize=(50,50))
sns.countplot(x='brand',data=df)
plt.xticks(rotation=90)
```

```python
data=[]
for i in df['brand'].unique():
    data.append([i,len(df[df['brand']==i])])
```

```python
data = pd.DataFrame(data,columns=['brands','Frequency'])
```

```python
data
```

```python
data.sort_values(by='Frequency',ascending=False)
```

# Car Industry Trend

```python
data=[]
for i in df['year'].unique():
    data.append([i,len(df[df['year']== i])])
```

```python
data = pd.DataFrame(data,columns=['Year','Frequency'])
```

```python
data
```

```python
data=data.sort_values(by = 'Year',ascending =True)
```

```python
plt.plot(data.Year,data.Frequency)
```

# Most Popular Color

```python
data=[]
for i in df['color'].unique():
    data.append([i,len(df[df['color']== i])])
```

```python
data = pd.DataFrame(data,columns=['Color','Frequency'])
```

```python
data = data.sort_values(by='Frequency',ascending=False)
```

```python
data
```

```python
data.plot(x='Color',y='Frequency',kind='bar',figsize=(15,5))
```

```python
sns.barplot(x='brand',y='price',data=df)
plt.xticks(rotation=90)
```

# Which Car brand is having high average price?

```python
df.groupby('brand')['price'].mean().sort_values(ascending=False).plot.bar(figsize=(10,5))
```

# Which state is having most expensive cars?

```python
plt.figure(figsize=(10,5))
df.groupby('state')['price'].mean().sort_values(ascending=False).plot.bar()
plt.show()
```
